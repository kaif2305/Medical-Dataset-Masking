```python
import csv
data = []

with open("Mdc_desc_hospital.csv", "r") as file:
    reader = csv.reader(file, delimiter=",") # 1
    for row in reader:
        data.append(" ".join(row)) # 2

with open("binary.csv", "w") as csvfile:
    csvwriter = csv.writer(csvfile)
    for row in data:
        csvwriter.writerow([row]) # 3
```
