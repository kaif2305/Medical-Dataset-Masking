```python
unq_char = np.unique(np.hstack(df.mdc_desc.str.split(','))).tolist()

for i in range(len(unq_char)):
  unq_char[i] = unq_char[i].rstrip()
  unq_char[i] = unq_char[i].lstrip()

print(unq_char)
```

```python
characters = ['ALCOHOL DRUG USE' , 'ALCOHOL DRUG INDUCED ORGANIC MENTAL DISEASES','BLOOD FORMING ORGANS','BREAST','CHILDBIRTH','CONNECTIVE TISSUE','DISEASES','DISORDERS','IMMUNOLOGICAL','METABOLIC','MOUTH','NEONATE CONDITIONS BEGAN IN PERINATAL PERIOD','NOSE','NUTRITIONAL','OTHER CONTACTS WITH HEALTH SERVICES','PANCREAS','PARASITIC DISEASES','POISONINGS','POORLY DIFFERENTIATED NEOPLASMS','SUBCUTANEOUS TISSUE','THE PUERPERIUM','THROAT','TOXIC EFFECTS OF DRUGS','URINARY TRACT','BLOOD','BURNS','CIRCULATORY SYSTEM','DIGESTIVE SYSTEM','EAR','ENDOCRINE','EYE','FACTORS ON HEALTH STATUS','FEMALE REPRODUCTIVE SYSTEM','HEPATOBILIARY SYSTEM','HUMAN IMMUNODEFICIENCY VIRUS INFECTIONS','INFECTIOUS','INJURIES','KIDNEY','MALE REPRODUCTIVE SYSTEM','MENTAL DISEASES','MULTIPLE SIGNFICANT TRAUMA','MUSCULOSKELETAL SYSTEM','MYELOPROLIFERATIVE DISEASES','NERVOUS SYSTEM','NEWBORNS','PREGNANCY','RESPIRATORY SYSTEM','SKIN','UNGROUPABLE']
```

```python
from sklearn import feature_extraction

coun_vect = feature_extraction.text.CountVectorizer(vocabulary=characters) 
count_matrix = coun_vect.fit_transform(df2['1'])

count_array = count_matrix.toarray()

dfFinal = pd.DataFrame(data=count_array,index=df2['1'],columns = coun_vect.get_feature_names())
```
