```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('stratified_hospital.csv')
```

```python
sc = StandardScaler()
x = df.iloc[:,3:4].values
df['Discharges'] = sc.fit_transform(x)
print(df['Discharges'])
```

```python
df.to_csv('After_Scaling_hospital.csv', index=False)

```
